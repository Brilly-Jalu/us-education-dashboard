# Data Preprocessing - Analisis Awal Dataset

Analisis awal untuk dataset Kmeans_assignment_data.csv, termasuk atribut, head, dan missing values.

---

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Project\Science Data\Visualization\Kmeans_assignment_data.csv")
```

```python
print("Atribut Data:")
print(df.columns.tolist())
```

```python
print("\nHead Data:")
print(df.head())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

---

```python
import matplotlib.pyplot as plt

# 1️⃣ Distribusi Graduation Rate
plt.figure()
plt.hist(df['Grad.Rate'], bins=20)
plt.title('Distribution of Graduation Rate')
plt.xlabel('Graduation Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2️⃣ Hubungan Tuition (Outstate) vs Graduation Rate
plt.figure()
plt.scatter(df['Outstate'], df['Grad.Rate'])
plt.title('Out-of-State Tuition vs Graduation Rate')
plt.xlabel('Out-of-State Tuition')
plt.ylabel('Graduation Rate')
plt.show()
```

```python
# 3️⃣ Full-time vs Part-time Undergraduate
plt.figure()
plt.scatter(df['F.Undergrad'], df['P.Undergrad'])
plt.title('Full-time vs Part-time Undergraduate Students')
plt.xlabel('Full-time Undergraduate')
plt.ylabel('Part-time Undergraduate')
plt.show()
```

```python
# 4️⃣ Student-Faculty Ratio vs Graduation Rate
plt.figure()
plt.scatter(df['S.F.Ratio'], df['Grad.Rate'])
plt.title('Student-Faculty Ratio vs Graduation Rate')
plt.xlabel('Student-Faculty Ratio')
plt.ylabel('Graduation Rate')
plt.show()
```

```python
# 5️⃣ Persentase Dosen PhD vs Graduation Rate
plt.figure()
plt.scatter(df['PhD'], df['Grad.Rate'])
plt.title('Faculty PhD Percentage vs Graduation Rate')
plt.xlabel('Percentage of Faculty with PhD')
plt.ylabel('Graduation Rate')
plt.show()
```
